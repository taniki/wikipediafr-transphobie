```python
import pandas as pd
from wekeypedia import WikipediaUser as User
```

```python
u = User(name="Sherwood6", lang="fr")
```

```python
revisions_full = u.fetch_contribs_full()
```

```python
len(revisions_full)
```

```python
df = (
    pd
    .DataFrame
    .from_records(
        [ 
            (

                {
                    'pageid': p['pageid'],
                    **rv,
                    'diff': rv['diff']['*'] if 'diff' in rv and '*' in rv['diff'] else '' 
                }
            )
            for p in revisions_full 
            for rv in p['revisions'] 
        ]
    )
    .set_index('revid')
)

df
```

```python
df.to_csv('../data/user-diffs/sherwood6.csv')
```
